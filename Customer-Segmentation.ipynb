## K- MEAN CLUSTERING

```python
data.columns
```

```python
x = data.iloc[:, [3,4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
k = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(x)
    k.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), k)
plt.show()
```

```python
model = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)
y_kmeans = model.fit_predict(x)
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
plt.figure(1, figsize=(15, 8))

# Cluster 1
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='magenta', label='Cluster 1')

# Cluster 2
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')

# Cluster 3
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='cyan', label='Cluster 3')

# Cluster 4
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='green', label='Cluster 4')

# Cluster 5
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='red', label='Cluster 5')

# Centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='black', label='Centroids')

plt.title('K Means Clustering Algorithm')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
kmeans.predict([[40,20]])
```

## Silhouette Coefficient-(Accuracy)

```python
from sklearn.metrics import silhouette_score

# Silhouette Score
silhouette = silhouette_score(X, y_kmeans)
print(f"Silhouette Score: {silhouette:.3f}")
```

```python
from tkinter import Tk, Label, Entry, Button
import joblib

# Load the model once
model = joblib.load("Customer-Segmentation.pkl")

# Function to handle prediction
def show_entry_fields():
    try:
        p1 = int(e1.get())
        p2 = int(e2.get())
        result = model.predict([[p1, p2]])
        cluster = result[0]c

        message = ""
        if cluster == 0:
            message = "Customers with medium annual income and medium annual spend"
        elif cluster == 1:
            message = "Customers with high annual income and high annual spend"
        elif cluster == 2:
            message = "Customers with high annual income and low annual spend"
        elif cluster == 3:
            message = "Customers with low annual income and low annual spend"
        elif cluster == 4:
            message = "Customers with low annual income and high annual spend"

        result_label.config(text=f"Cluster: {cluster}\n{message}")
    except ValueError:
        result_label.config(text="Please enter valid numbers.")

# GUI Setup
master = Tk()
master.title("Customer Segmentation using Machine Learning")
master.geometry("450x300")

Label(master, 
      text="Customer Segmentation using Machine Learning", 
      bg="black", 
      fg="white", 
      font=("Arial", 14)).grid(row=0, columnspan=2, pady=10)

Label(master, text="Annual Income (k$):").grid(row=1, column=0, pady=5)
Label(master, text="Spending Score (1-100):").grid(row=2, column=0, pady=5)

e1 = Entry(master)
e2 = Entry(master)
e1.grid(row=1, column=1, pady=5)
e2.grid(row=2, column=1, pady=5)

Button(master, text="Predict", command=show_entry_fields, bg="blue", fg="white").grid(row=3, columnspan=2, pady=10)

# Label to show the result
result_label = Label(master, text="", fg="green", font=("Arial", 12))
result_label.grid(row=4, columnspan=2, pady=10)

master.mainloop()
```
